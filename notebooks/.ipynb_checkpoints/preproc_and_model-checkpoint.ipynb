```python
from PIL import Image


def load_image( infilename ) :
    img = Image.open( infilename )
    img.load()
    data = np.asarray( img, dtype="int32" )
    return data
```

```python
import glob

norm_images = []
for f in glob.iglob("../data/train/NORMAL/*"):
    norm_images.append(load_image(f))
```
